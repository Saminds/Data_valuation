```python
import pickle
import matplotlib.pyplot as plt
```

```python
AFedsvg = open('./fmnist_cnn_100_AFedsvg.pkl', 'rb')
AFedsvg1 =  pickle.load(AFedsvg)
```

```python
fedavg = open('./fmnist_cnn_100_fedavg.pkl', 'rb')
fedavg1 =  pickle.load(fedavg)
```

```python
sfedavg = open('./fmnist_cnn_100_sfedavg.pkl', 'rb')
sfedavg1 =  pickle.load(sfedavg)
```

```python
import matplotlib.pyplot as plt

# Assuming AFedsvg1, fedavg1, sfedavg1 are your dictionaries containing 'test_accuracy' lists
# And assuming communication rounds are simply from 1 to 100

# Generate a list of communication rounds from 1 to 100
comm_rounds = list(range(1, 101))

# Plotting
plt.figure(figsize=(10, 6))

# Plot for AFedsvg1
plt.plot(comm_rounds, AFedsvg1['test_accuracy'], label='AFedsvg1')

# Plot for fedavg1
plt.plot(comm_rounds, fedavg1['test_accuracy'], label='FedAvg1')

# Plot for sfedavg1
plt.plot(comm_rounds, sfedavg1['test_accuracy'], label='SFedAvg1')

# Adding titles and labels
plt.title('Test Accuracy vs. Communication Rounds')
plt.xlabel('Communication Rounds')
plt.ylabel('Test Accuracy')
plt.legend()

# Show the plot
plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt
fedavg1 = open('./fmnist_cnn_100_Clients[100]_FedAvg_noise7.pkl', 'rb')
ReverseGradient =  pickle.load(fedavg1)

GradientNoise1 = open('./fmnist_cnn_100_Clients[100]_5FedAvg.pkl', 'rb')
GradientNoise =  pickle.load(GradientNoise1)

RandomAttack1 = open('./fmnist_cnn_100_Clients[100]_6_FedAvg.pkl', 'rb')
RandomAttack =  pickle.load(RandomAttack1)
# Assuming AFedsvg1, fedavg1, sfedavg1 are your dictionaries containing 'test_accuracy' lists
# And assuming communication rounds are simply from 1 to 100

# Generate a list of communication rounds from 1 to 100
comm_rounds = list(range(1, 101))

# Plotting
plt.figure(figsize=(10, 6))

# Plot for AFedsvg1
plt.plot(comm_rounds, ReverseGradient['test_accuracy'], 'g+-',label='ReverseGradient')
plt.plot(comm_rounds, GradientNoise['test_accuracy'], 'ro-',label='GradientNoise')
plt.plot(comm_rounds, RandomAttack['test_accuracy'], 'b^-',label='RandomAttack')

# Adding titles and labels
plt.title('Test Accuracy vs. Communication Rounds')
plt.xlabel('Communication Rounds')
plt.ylabel('Test Accuracy')
plt.legend()

# Show the plot
plt.show()
```
